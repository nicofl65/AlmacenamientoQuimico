```python
#------------------------ PANELES FOTOVOLTAICOS -------------------------
Eta_PV = 0.23     #eficiencia de los paneles fotovoltaicos
Area_PV_unit = 2.583252000      #Área unitaria de los paneles fotovoltaicos, m2
Precio_PV = 109.59      #precio unitario placas PV, €
OMC_PV_unit = 10 #Costes de operación y mantenimiento de una placa fotovoltaica (anual), €/año/ud
PV_Units = 637          #número de paneles
Area_PV_TOT = PV_Units * Area_PV_unit     #área total de paneles, m2

Pot_PV = Rad_global * Eta_PV * Area_PV_TOT      #potencia fotovoltaica horaria, kW
Pot_PV_TOT = np.sum(Pot_PV)     #sumatorio de potencia fotovoltaica anual (energía), kWh

#-------------------------- CONSUMO ----------------------------
Consumo_TOT = np.sum(Consumo)     #consumo anual, kWh

Precio_Grid=0.12      #Precio de la electricidad de red, €/kWh
Coste_Grid=Precio_Grid*Consumo_TOT      #Coste anual de electricidad de red, €

#--------------------------- BATERÍAS -------------------------------
Precio_Bat = 9242     #precio unitario baterías, €
OMC_Bat_unit = 0      #Costes de operación y mantenimiento de una batería (anual), €/año/ud
Excedente = Pot_PV - Consumo      #potencia que NO queda cubierta por los paneles (horario), kW

Carga_Bat = np.where(Excedente > 0, Excedente, 0)     #carga de la batería con excedente de PV (horario), kW

Descarga_Bat = np.where(Excedente <0, Excedente, 0)     #descarga de batería (valores negativos, horario), kW

Pot_bat=15.4      #potencia máxima de cada batería, kWh
Bat_Units=40      #número de baterías

Pot_max_bat=Bat_Units*Pot_bat     #Potencia máxima del conjunto de baterías, kW

N = 8760
Estado_Bat = np.zeros(N)

Estado_Bat[0] = 600   # valor inicial deseado, kWh

for t in range(1, N):     #Cálculo del estado de las baterías, siempre tiene que ser mayor que cero y menor que el máximo de potencia
    Estado_Bat[t] = max(0, min(Estado_Bat[t-1] + Carga_Bat[t] + Descarga_Bat[t],Pot_max_bat))

Pot_Bat_TOT=np.sum(Carga_Bat)     #Energía total cargada en las baterías en un año

# ---------------------- COSTES ------------------------
CAPEX_PV=PV_Units*Precio_PV
# CAPEX_Bat=Bat_Units*Precio_Bat
# CAPEX_TOT=CAPEX_PV+CAPEX_Bat

OMC_PV=OMC_PV_unit*PV_Units
# OMC_Bat=OMC_Bat_unit*Bat_Units
# OMC_TOT=OMC_PV+OMC_Bat


# Payback = CAPEX_TOT/(Coste_Grid-OMC_TOT)

print("Energía fotovoltaica anual, kWh: ", f"{Pot_PV_TOT:.2f}")
print("Energía consumida anual, kWh: ", f"{Consumo_TOT:.2f}")
print("Energía de carga de las baterías en un año, kWh: ", f"{Pot_Bat_TOT:.2f}")
# print("CAPEX total, €: ", CAPEX_TOT)
# print("Costes O&M totales, €: ", OMC_TOT)
# print("Payback: ", f"{Payback:.2f}")
```

```python
def calcular_autoconsumo(capacidad_total, eff_carga, eff_descarga, dod):
    n = len(Pot_PV)
    estado_bat = 0
    energia_autoconsumida = 0
    capacidad_util = capacidad_total * dod

    for t in range(n):
        autoconsumo_directo = min(Pot_PV[t], Consumo[t])
        excedente = Pot_PV[t] - autoconsumo_directo
        deficit = Consumo[t] - autoconsumo_directo

        espacio_libre = capacidad_util - estado_bat
        carga_posible = min(excedente, espacio_libre / eff_carga)
        estado_bat += carga_posible * eff_carga

        descarga_necesaria = deficit / eff_descarga
        descarga_real = min(descarga_necesaria, estado_bat)
        estado_bat -= descarga_real
        energia_autoconsumida += (autoconsumo_directo + descarga_real * eff_descarga)

    return energia_autoconsumida

# Escenarios basados en los enlaces proporcionados:
# 1. BYD LFP (15.4 kWh)
# 2. FZSoNick Sodio (9.6 kWh)
# 3. Saft Ni-Cd (2.0 kWh aprox por bloque industrial)
# 4. Zenaji LTO (1.93 kWh por módulo Aeon)

escenarios_bat = [
    # (Capacidad_kWh, Precio_€, "Modelo (Tecnología)", E_carga, E_descarga, DOD)
    (15.4, 6075, "BYD LVL 15.4 (LFP)", 0.96, 0.96, 0.90),
    (9.6, 9250, "FZSoNick 48TL200 (Sodio alta T)", 0.90, 0.90, 1.00),
    (2.0, 1500, "Saft Sunica.plus (Ni-Cd)", 0.95, 0.95, 1.00),     # Capacidad/precio de bloque modular representativo
   (1.93, 2200, "Zenaji Aeon (LTO)", 0.98, 0.98, 1.0)]

print(f"Energía Solar Total Producida: {Pot_PV_TOT:.2f} kWh\n")

for cap_unit, precio, nombre, eff_c, eff_d, dod in escenarios_bat:
    unidades = 0
    autoconsumo_actual = 0
    objetivo = Pot_PV_TOT * 0.999
    prev_autoconsumo = 0

    while autoconsumo_actual < objetivo and unidades < 1000:
        unidades += 1
        autoconsumo_actual = calcular_autoconsumo(unidades * cap_unit, eff_c, eff_d, dod)
        if unidades > 1:
            if abs(autoconsumo_actual - prev_autoconsumo) < 0.1:
                unidades -= 1
                break
        prev_autoconsumo = autoconsumo_actual

    porcentaje = (autoconsumo_actual / Pot_PV_TOT) * 100
    print(f"--- {nombre} ({cap_unit} kWh) ---")
    print(f"Eficiencia Ciclo: {eff_c*eff_d*100:.1f}% | DoD: {dod*100:.0f}%")
    print(f"Unidades necesarias: {unidades}")
    print(f"Capacidad instalada: {unidades * cap_unit:.2f} kWh")
    print(f"Autoconsumo alcanzado: {porcentaje:.2f}% de la energía solar")
    print(f"Inversión en baterías: {unidades * precio:,.2f} €\n")
```

```python
def calcular_sistema_hibrido(capacidad_bat_kwh, precio_gen_kwh):
    n = len(Pot_PV)
    estado_bat = 0
    energia_solar_directa = 0
    energia_bat_descargada = 0
    energia_generador = 0

    for t in range(n):
        autoconsumo_directo = min(Pot_PV[t], Consumo[t])
        excedente = Pot_PV[t] - autoconsumo_directo
        deficit = Consumo[t] - autoconsumo_directo

        # Carga batería
        espacio_libre = capacidad_bat_kwh - estado_bat
        carga = min(excedente, espacio_libre)
        estado_bat += carga

        # Descarga batería
        descarga = min(deficit, estado_bat)
        estado_bat -= descarga

        # El resto del déficit lo cubre el generador
        pendiente = deficit - descarga
        energia_generador += pendiente

        energia_solar_directa += autoconsumo_directo
        energia_bat_descargada += descarga

    return energia_generador, (energia_solar_directa + energia_bat_descargada)

# Parámetros del generador
precio_combustible_kwh = 0.35  # Coste estimado por kWh generado con diesel/gas

# Vamos a comparar el Modelo 4 (el más eficiente) con diferentes niveles de almacenamiento
cap_unit_m4 = 213.0
precio_unit_m4 = 80000

casos_unidades = [0, 1, 2, 3] # De 0 baterías a 3 (el máximo técnico anterior)
inversion_bat = []
coste_operacion_gen = []
autoconsumo_total = []

for u in casos_unidades:
    cap_tot = u * cap_unit_m4
    gen_kwh, auto_kwh = calcular_sistema_hibrido(cap_tot, precio_combustible_kwh)

    inversion_bat.append(u * precio_unit_m4)
    coste_operacion_gen.append(gen_kwh * precio_combustible_kwh)
    autoconsumo_total.append(auto_kwh)

# Visualización
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.bar(casos_unidades, inversion_bat, label='Inversión Baterías (CAPEX)', color='navy', alpha=0.6)
ax1.set_xlabel('Unidades de Batería (Modelo 4 - 213kWh)')
ax1.set_ylabel('CAPEX Baterías (€)', color='navy')

ax2 = ax1.twinx()
ax2.plot(casos_unidades, coste_operacion_gen, 'r-o', label='Coste Anual Generador (OPEX)')
ax2.set_ylabel('Coste Operativo Generador (€/año)', color='red')

plt.title('Trade-off: Inversión en Baterías vs Coste de Generación Auxiliar')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

for i, u in enumerate(casos_unidades):
    print(f"Con {u} unidades: Inversión {inversion_bat[i]:,.0f}€ | Gasto anual Generador: {coste_operacion_gen[i]:,.2f}€")
```

```python
import numpy_financial as npf

# Parámetros financieros
años = 10
tasa_descuento = 0.05

van_resultados = []

# El escenario base (referencia) sería el coste de pagar el 100% de la red sin paneles ni baterías
coste_anual_solo_red = Consumo_TOT * Precio_Grid

print(f"Análisis de VAN a {años} años (Tasa: {tasa_descuento*100}%)\n")

for i, u in enumerate(casos_unidades):
    # Inversión inicial (Baterías + Paneles)
    inversion_inicial = inversion_bat[i] + CAPEX_PV

    # Costes anuales (O&M + Generador)
    om_anual = OMC_PV + (u * OMC_Bat_unit)
    gasto_generador = coste_operacion_gen[i]
    coste_total_anual = om_anual + gasto_generador

    # Ahorro anual comparado con no tener nada (solo red)
    # Nota: El sistema PV + Bat + Gen cubre el 100% del consumo
    ahorro_anual = coste_anual_solo_red - coste_total_anual

    # Flujos de caja: [Inversión (negativa), ahorro1, ahorro2, ..., ahorro10]
    flujos = [-inversion_inicial] + [ahorro_anual] * años

    # Cálculo del VAN
    van = npf.npv(tasa_descuento, flujos)
    van_resultados.append(van)

    print(f"Escenario {u} bat ({u*cap_unit_m4} kWh):")
    print(f"  - Inversión Inicial: {inversion_inicial:,.2f} €")
    print(f"  - Ahorro Neto Anual: {ahorro_anual:,.2f} €")
    print(f"  - VAN: {van:,.2f} €")
    print("-" * 30)

# Gráfica del VAN
plt.figure(figsize=(10, 6))
plt.bar([f"{u} Unidades" for u in casos_unidades], van_resultados, color=['gray', 'blue', 'green', 'orange'])
plt.axhline(0, color='black', linewidth=0.8)
plt.ylabel('Valor Actual Neto (€)')
plt.title('Comparativa de VAN a 10 años por número de baterías')
for i, v in enumerate(van_resultados):
    plt.text(i, v + (max(van_resultados)*0.05 if v>0 else -max(van_resultados)*0.1), f'{v:,.0f}€', ha='center', fontweight='bold')
plt.show()
```

```python
def verificar_cobertura_total(capacidad_total, eff_c, eff_d, dod):
    n = len(Pot_PV)
    estado_bat = capacidad_total * dod # Empezamos con batería llena para el ciclo anual
    capacidad_util = capacidad_total * dod
    
    for t in range(n):
        autoconsumo_directo = min(Pot_PV[t], Consumo[t])
        excedente = Pot_PV[t] - autoconsumo_directo
        deficit = Consumo[t] - autoconsumo_directo

        # Carga
        espacio_libre = capacidad_util - estado_bat
        carga_posible = min(excedente, espacio_libre / eff_c)
        estado_bat += carga_posible * eff_c

        # Descarga
        descarga_necesaria = deficit / eff_d
        if descarga_necesaria > estado_bat:
            return False # La batería se agota
        estado_bat -= descarga_necesaria
    
    return True

print("Optimización de Unidades Mínimas para evitar Estado_Bat = 0:\n")

for cap_unit, precio, nombre, eff_c, eff_d, dod in escenarios_bat:
    unidades = 1
    cubierto = False
    
    # Búsqueda incremental del mínimo de unidades
    while not cubierto and unidades < 2000:
        if verificar_cobertura_total(unidades * cap_unit, eff_c, eff_d, dod):
            cubierto = True
        else:
            unidades += 1
    
    if cubierto:
        coste_inst = unidades * precio
        print(f"--- {nombre} ---")
        print(f"  Unidades mínimas: {unidades}")
        print(f"  Capacidad total: {unidades * cap_unit:.2f} kWh")
        print(f"  Coste de instalación: {coste_inst:,.2f} €\n")
    else:
        print(f"--- {nombre} --- No se encontró solución técnica con menos de 2000 unidades.\n")
```


## 2) Simulación común para comparar las 4 baterías

La comparación usa la misma capacidad nominal de referencia que ya tenías calculada arriba (`Pot_max_bat`). Así se compara tecnología contra tecnología sin cambiar la base del sistema.


```python

# ============================================================
# 2) Función de simulación PV + batería + respaldo
# ============================================================
def simular_pv_bateria_diesel(
    capacidad_nominal_kWh,
    eff_rt,
    dod,
    diesel_power_kW=np.inf,
    soc_inicial_frac=0.0
):
    """
    Simula una micro-red horaria con PV, batería y generador diésel.

    Pot_PV y Consumo se interpretan como kWh por hora, equivalentes a kW medios horarios.
    La batería se modela con eficiencia round-trip repartida simétricamente entre carga y descarga.
    """
    eff_c = np.sqrt(float(eff_rt))
    eff_d = np.sqrt(float(eff_rt))
    capacidad_util = float(capacidad_nominal_kWh) * float(dod)
    soc = capacidad_util * float(soc_inicial_frac)

    energia_solar_directa = 0.0
    energia_cargada_bateria = 0.0
    energia_descargada_bateria = 0.0
    energia_diesel = 0.0
    energia_vertida = 0.0
    energia_no_suministrada = 0.0

    soc_hist = np.zeros(len(Consumo))
    diesel_hist = np.zeros(len(Consumo))
    ens_hist = np.zeros(len(Consumo))
    deficit_antes_diesel_hist = np.zeros(len(Consumo))

    for t in range(len(Consumo)):
        generacion = max(float(Pot_PV[t]), 0.0)
        demanda = max(float(Consumo[t]), 0.0)

        directo = min(generacion, demanda)
        excedente = generacion - directo
        deficit = demanda - directo
        energia_solar_directa += directo

        # Carga de batería con excedentes FV
        espacio_util = capacidad_util - soc
        carga_entrada = min(excedente, espacio_util / eff_c) if eff_c > 0 else 0.0
        soc += carga_entrada * eff_c
        energia_cargada_bateria += carga_entrada
        energia_vertida += max(excedente - carga_entrada, 0.0)

        # Descarga de batería para cubrir déficit
        descarga_almacen = min(deficit / eff_d, soc) if eff_d > 0 else 0.0
        energia_entregada_bat = descarga_almacen * eff_d
        soc -= descarga_almacen
        energia_descargada_bateria += energia_entregada_bat

        pendiente = max(deficit - energia_entregada_bat, 0.0)
        deficit_antes_diesel_hist[t] = pendiente

        # Generador diésel limitado por potencia horaria
        diesel_t = min(pendiente, diesel_power_kW)
        energia_diesel += diesel_t
        diesel_hist[t] = diesel_t

        ens_t = max(pendiente - diesel_t, 0.0)
        energia_no_suministrada += ens_t
        ens_hist[t] = ens_t

        soc_hist[t] = soc

    return {
        "solar_directa_kWh": energia_solar_directa,
        "bateria_carga_kWh": energia_cargada_bateria,
        "bateria_descarga_kWh": energia_descargada_bateria,
        "diesel_kWh": energia_diesel,
        "vertido_kWh": energia_vertida,
        "ens_kWh": energia_no_suministrada,
        "autoconsumo_total_kWh": energia_solar_directa + energia_descargada_bateria,
        "cobertura_renovable_pct": 100 * (energia_solar_directa + energia_descargada_bateria) / energia_consumo_anual if energia_consumo_anual else np.nan,
        "soc_min_kWh": float(np.min(soc_hist)) if len(soc_hist) else 0.0,
        "soc_max_kWh": float(np.max(soc_hist)) if len(soc_hist) else 0.0,
        "diesel_hist": diesel_hist,
        "ens_hist": ens_hist,
        "soc_hist": soc_hist,
        "deficit_antes_diesel_hist": deficit_antes_diesel_hist,
        "potencia_diesel_necesaria_kW": float(np.max(deficit_antes_diesel_hist)) if len(deficit_antes_diesel_hist) else 0.0,
    }
```

```python

# ============================================================
# 3) Comparativa de 4 tecnologías con la capacidad ya definida
# ============================================================
comparativa = []

for _, bat in baterias.iterrows():
    sim = simular_pv_bateria_diesel(
        capacidad_nominal_kWh=capacidad_referencia_kwh,
        eff_rt=bat["eff_rt"],
        dod=bat["dod"],
        diesel_power_kW=np.inf,   # aquí se calcula cuánto haría falta antes de limitar el grupo
        soc_inicial_frac=0.0
    )

    capex_bateria = capacidad_referencia_kwh * bat["capex_eur_kWh"]
    opex_bateria = capacidad_referencia_kwh * bat["opex_eur_kWh_year"] if pd.notna(bat["opex_eur_kWh_year"]) else np.nan

    comparativa.append({
        "tecnologia": bat["tecnologia"],
        "nombre": bat["nombre"],
        "capacidad_nominal_kWh": capacidad_referencia_kwh,
        "capacidad_util_kWh": capacidad_referencia_kwh * bat["dod"],
        "eff_rt": bat["eff_rt"],
        "DoD": bat["dod"],
        "CAPEX_bateria_eur": capex_bateria,
        "OPEX_bateria_eur_year": opex_bateria,
        "diesel_necesario_sin_limite_kWh_year": sim["diesel_kWh"],
        "potencia_diesel_necesaria_kW": sim["potencia_diesel_necesaria_kW"],
        "vertido_kWh_year": sim["vertido_kWh"],
        "cobertura_renovable_pct": sim["cobertura_renovable_pct"],
        "seguridad_score": bat["seguridad_score_calculado"],
        "datos_incompletos": bat["datos_incompletos"],
    })

comparativa_baterias = pd.DataFrame(comparativa)

# Métrica de coste parcial: batería + OPEX si existe. El diésel se añade en el siguiente bloque.
comparativa_baterias["coste_anual_bateria_conocido_eur"] = comparativa_baterias["OPEX_bateria_eur_year"]

# Score transparente: menor CAPEX, menor necesidad de diésel, mayor cobertura renovable y mayor seguridad.
def normalizar_positivo(s):
    s = pd.Series(s, dtype=float)
    if s.max() == s.min():
        return pd.Series(np.ones(len(s)), index=s.index)
    return (s - s.min()) / (s.max() - s.min())

def normalizar_negativo(s):
    return 1 - normalizar_positivo(s)

comparativa_baterias["score_capex"] = normalizar_negativo(comparativa_baterias["CAPEX_bateria_eur"])
comparativa_baterias["score_diesel"] = normalizar_negativo(comparativa_baterias["diesel_necesario_sin_limite_kWh_year"])
comparativa_baterias["score_cobertura"] = normalizar_positivo(comparativa_baterias["cobertura_renovable_pct"])
comparativa_baterias["score_seguridad"] = comparativa_baterias["seguridad_score"] / 5
comparativa_baterias["penalizacion_datos"] = np.where(comparativa_baterias["datos_incompletos"], 0.10, 0.0)

comparativa_baterias["score_total"] = (
    0.30 * comparativa_baterias["score_capex"] +
    0.25 * comparativa_baterias["score_diesel"] +
    0.25 * comparativa_baterias["score_cobertura"] +
    0.20 * comparativa_baterias["score_seguridad"] -
    comparativa_baterias["penalizacion_datos"]
)

comparativa_baterias = comparativa_baterias.sort_values("score_total", ascending=False).reset_index(drop=True)

tecnologia_optima = comparativa_baterias.loc[0, "tecnologia"]
bateria_optima = baterias[baterias["tecnologia"] == tecnologia_optima].iloc[0]

print(f"Tecnología óptima según el score calculado: {tecnologia_optima}")
display(comparativa_baterias)

plt.figure(figsize=(10, 5))
plt.bar(comparativa_baterias["tecnologia"], comparativa_baterias["score_total"])
plt.ylabel("Score total")
plt.title("Ranking multicriterio de tecnologías de batería")
plt.show()
```


## 5) capacidad óptima de LTO según precio del diésel

```python

# ============================================================
# 6) Sensibilidad conjunta: capacidad de batería óptima vs precio del diésel
# ============================================================
def factor_recuperacion_capital(tasa, years):
    if tasa == 0:
        return 1 / years
    return tasa * (1 + tasa) ** years / ((1 + tasa) ** years - 1)

# Parámetros financieros editables.
HORIZONTE_YEARS = 20
TASA_DESCUENTO = 0.05
CRF = factor_recuperacion_capital(TASA_DESCUENTO, HORIZONTE_YEARS)

# Se toma LTO si existe, porque es la tecnología defendida en el informe para la configuración final.
if "LTO" in set(baterias["tecnologia"]):
    bat_sens = baterias[baterias["tecnologia"] == "LTO"].iloc[0]
else:
    bat_sens = bateria_optima

cap_min = 0
cap_max = max(capacidad_referencia_kwh * 1.5, capacidad_referencia_kwh + 300)
capacidades_test = np.arange(cap_min, cap_max + 1, 50)
precios_test = np.array([
    DIESEL_BASE_PRICE_EUR_L * 0.70,
    DIESEL_BASE_PRICE_EUR_L,
    DIESEL_BASE_PRICE_EUR_L * 1.30,
])

filas_opt_cap = []
for precio_diesel in precios_test:
    for cap in capacidades_test:
        sim = simular_pv_bateria_diesel(
            capacidad_nominal_kWh=cap,
            eff_rt=float(bat_sens["eff_rt"]),
            dod=float(bat_sens["dod"]),
            diesel_power_kW=DIESEL_POWER_KW,
            soc_inicial_frac=0.0
        )
        capex_bat = cap * float(bat_sens["capex_eur_kWh"])
        opex_bat = cap * float(bat_sens["opex_eur_kWh_year"]) if pd.notna(bat_sens["opex_eur_kWh_year"]) else 0.0
        litros = sim["diesel_kWh"] * DIESEL_L_PER_KWH
        coste_diesel = litros * precio_diesel + sim["diesel_kWh"] * DIESEL_OPEX_EUR_KWH

        # Coste anual equivalente del sistema de respaldo: batería + diésel.
        coste_anual_equivalente = capex_bat * CRF + opex_bat + coste_diesel

        filas_opt_cap.append({
            "precio_diesel_eur_L": precio_diesel,
            "capacidad_bateria_kWh": cap,
            "diesel_kWh_year": sim["diesel_kWh"],
            "ens_kWh_year": sim["ens_kWh"],
            "coste_anual_equivalente_eur_year": coste_anual_equivalente,
            "capex_bateria_eur": capex_bat,
            "opex_bateria_eur_year": opex_bat,
            "coste_diesel_eur_year": coste_diesel,
        })

opt_capacidad_lto = pd.DataFrame(filas_opt_cap)

# Para elegir óptimo realista, penalizamos soluciones con energía no suministrada.
# Si quieres permitir ENS, elimina esta línea de filtrado.
opt_validas = opt_capacidad_lto[opt_capacidad_lto["ens_kWh_year"] <= 1e-6].copy()
if opt_validas.empty:
    print("Aviso: con 48 kW de diésel todas las combinaciones dejan algo de energía no suministrada. Se muestra el menor coste sin filtrar ENS.")
    opt_validas = opt_capacidad_lto.copy()

idx = opt_validas.groupby("precio_diesel_eur_L")["coste_anual_equivalente_eur_year"].idxmin()
capacidad_optima_segun_diesel = opt_validas.loc[idx].sort_values("precio_diesel_eur_L").reset_index(drop=True)

display(capacidad_optima_segun_diesel)

plt.figure(figsize=(10, 5))
for precio_diesel, dfp in opt_capacidad_lto.groupby("precio_diesel_eur_L"):
    plt.plot(
        dfp["capacidad_bateria_kWh"],
        dfp["coste_anual_equivalente_eur_year"],
        marker="o",
        label=f"Diésel {precio_diesel:.2f} €/L"
    )
plt.xlabel("Capacidad nominal batería LTO (kWh)")
plt.ylabel("Coste anual equivalente (€/año)")
plt.title("Capacidad óptima de LTO según precio del diésel")
plt.legend()
plt.show()
```
